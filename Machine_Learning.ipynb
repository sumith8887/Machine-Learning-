#### What is a Model in ML?

* A model in ML is a mathematical representation of the relationship between input and output





##### Components of a Model

* Parameters
* Hyperparameters
* Model Structure



###### How does a Model Learn?

* A model learns by adjusting its parameters based on the training data.

  During training :

  * bad prediction --> adjust parameters
  * better prediction --> Keep adjusting
  * finally --> model learns the best parameters

  ###### 

  ###### Loss(Error) function :

* A loss function tells us how wrong the model's prediction is.

  * small loss --> good prediction
  * Large loss --> bad prediction

  

  ###### Optimization Goal :

* The goal of training is Reducing the loss as much as possible.
  We do this By:

  * adjusting the parameters
  * Repeating this process

### Polynomial Regression

Polynomial regression is a form of regression analysis in which the relationship between the independent variable x and dependent variable y is modeled as an nth degree polynomial of x. That is, if your dataset holds the characteristic of being curved when plotted in the graph, then polynomial regression is used to fit the data more accurately than linear regression.

f(x) = a0+a1x+a2x^2+a3x^3+.....+anx^n
